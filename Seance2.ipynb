```r
donnees_detaillees = data.frame(
                            Age = c('18-24', '18-24', '25-34', '25-34',  '35-44', '35-44', '45-54', '45-54', 
                                    '55-64', '55-64', '65-74', '65-74', '75+', '75+'),
                            Statut = c('F', 'NF','F', 'NF', 'F', 'NF', 'F', 'NF', 'F', 'NF', 'F', 'NF', 'F', 'NF'),
                            Effectif_deces = c(2, 1, 3, 5, 14, 7, 27, 12, 51, 40, 29, 101, 13, 64),
                            Effectif_vie = c(53, 61, 121, 152, 95, 114, 103, 66, 64, 81, 7, 28, 0, 0))
donnees_detaillees
```

```r
prop_deces_par_age_statut = donnees_detaillees$Effectif_deces/(donnees_detaillees$Effectif_deces+ donnees_detaillees$Effectif_vie)
donnees_detaillees <- data.frame(donnees_detaillees, prop_deces_par_age_statut)
donnees_detaillees
```

```r
donnees_detaillees_prop <- reshape(donnees_detaillees[ , 1:2-5],  #On prend les deux premières colonnes (qui classent) et la dernière avec les proportions
                     idvar = "Statut",
                     timevar = "Age",
                     direction = "wide")
row.names(donnees_detaillees_prop) <- donnees_detaillees_prop$Statut
donnees_detaillees_prop <- donnees_detaillees_prop[ , 2:ncol(donnees_detaillees_prop)]
colnames(donnees_detaillees_prop) <- c('18-24', '25-34',  '35-44', '45-54', '55-64', '65-74', '75+')
donnees_detaillees_prop <- as.matrix(donnees_detaillees_prop)
donnees_detaillees_prop
```

```r
options(repr.plot.width = 16, repr.plot.height = 8) # Ne pas s'occuper de cette ligne, c'est juste un paramètre d'affichage

barplot(height = donnees_detaillees_prop,
        beside = TRUE, 
        main="Taux de décès par âge et selon le statut (F ou NF)",
        ylab="Proportion", col=c("gold","red"),
        legend = c("Fumeuses", "Non-fumeuses"), args.legend=list(x="topleft"))
```

Mais quelle est la proportion de fumeuses/non-fumeuses interrogées par âge? 

```r
effectif_F_par_age = c(2+32, 3+121, 14+95, 27+103, 51+64, 29+7, 13)
effectif_NF_par_age = c(1+61, 5+152, 7+114, 12+66, 40+81, 101+28, 64)
effectif_total_par_age = effectif_F_par_age+ effectif_NF_par_age
```

```r
prop_F_par_age = effectif_F_par_age/effectif_total_par_age
prop_NF_par_age = effectif_NF_par_age/effectif_total_par_age

prop_F_par_age
prop_NF_par_age
```

```r
options(repr.plot.width = 16, repr.plot.height = 8) # Ne pas s'occuper de cette ligne, c'est juste un paramètre d'affichage

barplot(t(matrix(c(prop_F_par_age, prop_NF_par_age), nrow= 7, ncol=2)), 
        main="Part des fumeuses/non-fumeuses interrogées",
        names.arg=c('18-24', '25-34',  '35-44', '45-54', '55-64', '65-74', '75+'),
        ylab="Proportions", col=c("gold","red"),
        legend = c("Fumeuses", "Non-fumeuses"), args.legend=list(x="bottomright"))

abline(h=0.5, col='blue')
```
